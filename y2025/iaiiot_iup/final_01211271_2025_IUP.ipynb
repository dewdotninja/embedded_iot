````python
class QLearningAgent:
    ACTION_NAMES = ["LEFT", "DOWN", "RIGHT", "UP"]
    ROWS = 8
    COLS = 8

    def __init__(
        self, state_shape=(8,8), learning_rate=0.1, discount_factor=0.9, epsilon=0.1
    ):
        # Q[row, col, action] = value of taking that action in that state
        self.q_table = np.zeros(state_shape + (4,))  # 4 actions
        self.learning_rate = learning_rate  # α
        self.discount_factor = discount_factor  # γ
        self.epsilon = epsilon  # ε
    def setepsilon(self,epsilon):
        self.epsilon = epsilon 
    def choose_action(self, state):
        """ε-greedy: sometimes explore, otherwise exploit."""
        if random.uniform(0, 1) < self.epsilon:
            return random.randint(0, 3)  # Explore (random action)
        return int(np.argmax(self.q_table[state]))  # Exploit (best known action)

    def update_q_value(self, state, action, reward, next_state, done):
        """
        Tabular Q-learning update:
        Q(s,a) ← Q(s,a) + α [ r + γ max_a' Q(s',a') - Q(s,a) ]
        If next_state is terminal, we use just r as the target.
        """
        current_q = self.q_table[state][action]

        if done:
            target = reward
        else:
            max_future_q = np.max(self.q_table[next_state])
            target = reward + self.discount_factor * max_future_q

        # write one line of code to adjust the Q-table here
        self.q_table[state][action] = ?
        

    def get_q_table(self): # get q-table
        return self.q_table

    def print_q_table(self, iteration: int | None = None, precision: int = 3) -> None:
        """
        Print Q-table as a 36x4 matrix:
        - Rows   = states (x, y)
        - Cols   = actions (UP, RIGHT, DOWN, LEFT)
        """

        if iteration is not None:
            print(f"\n=== Q-table (iteration {iteration}) ===")

        # Header
        header = "State (x,y) | " + " | ".join(f"{a:^8}" for a in self.ACTION_NAMES)
        print(header)
        print("-" * len(header))

        rows, cols, _ = self.q_table.shape

        for i in range(rows):  # row (y)
            for j in range(cols):  # col (x)
                state_label = f"({j},{i})".ljust(10)

                values = [f"{self.q_table[i, j, a]:>{8}.{precision}f}" for a in range(4)]

                print(f"{state_label} | " + " | ".join(values))

        print("=" * len(header))
````

````python

ARROWS = ["←", "↓", "→", "↑"]

def plot_grid(agent=None, show_coords=False,show_q = False):
    """
    If agent is provided, show best-action arrows in non-terminal cells.
    Otherwise, just show S / G / X.
    """
    ROWS = 8
    COLS = 8
    cmap = ListedColormap(["lightblue", "red", "green"])

    viz = np.zeros_like(rmat_ex, dtype=np.int32)

    for i in range(ROWS):
        for j in range(COLS):
            if rmat_ex[i,j] > 0: # positive reward
                viz[i,j] = 2  # green
            elif rmat_ex[i,j] < 0: # negative reward
                viz[i,j] = 1 # red

    plt.figure(figsize=(8,8))
    plt.imshow(viz, cmap=cmap)

    for i in range(ROWS):
        for j in range(COLS):
            if rmat_ex[i,j] > 0: # positive reward
                plt.text(j,i,str(rmat_ex[i,j]),ha="center", va="center", fontweight="bold")
            elif rmat_ex[i,j] < 0: # negative reward
                plt.text(j,i,str(rmat_ex[i,j]),ha="center", va="center", fontweight="bold")
            else:
                if agent is not None:
                    a = int(np.argmax(agent.q_table[i, j]))
                    plt.text(j, i, ARROWS[a], ha="center", va="center", color="green",fontsize=14)
                    if show_q:
                        q_str = (str(np.round(agent.q_table[i][j][3],2))+"\n\n\n"+ str(np.round(agent.q_table[i][j][0],2)) 
                            + "      " + str(np.round(agent.q_table[i][j][2],2)) +"\n\n"+ str(np.round(agent.q_table[i][j][1],2)))
                        plt.text(j,i,q_str,ha="center",va="center",fontsize=8)
                elif show_coords:
                    plt.text(j, i, f"({j},{i})", ha="center", va="center")



    # Start marker
    si, sj = 0,0
    plt.text(sj, si, "Start", ha="center", va="bottom", fontweight="bold")

    # ----- ticks & grid lines -----
    plt.xticks(range(COLS))
    plt.yticks(range(ROWS))

    # Minor ticks at cell boundaries
    plt.gca().set_xticks(np.arange(-0.5, COLS, 1), minor=True)
    plt.gca().set_yticks(np.arange(-0.5, ROWS, 1), minor=True)

    # Draw black grid lines
    plt.grid(which="minor", color="black", linestyle="-", linewidth=1)
    plt.grid(False)  # disable major grid

    plt.title("Frozen-Lake 8x8" + (" (policy)" if agent is not None else ""))
    # plt.savefig("6x6_grid_world.png", dpi=600)
    plt.show()
````
