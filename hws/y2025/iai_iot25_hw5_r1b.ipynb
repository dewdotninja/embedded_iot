The requirements for this embedded system are (Requirement 1 - 3 are the same as in HW4, except that it must be programmed using micropython. Requirement 4 is slightly modified to add a single random number)

1. The target is the distance from a wall. It can be adjusted via command "target = xxx" where xxx is the distance from the wall in range 50 - 300 cm. The command intepreter must be friendly; i.e., the user can type "target = 50", "target=120", "TARGET  =   100.0", or "tArGeT=75" and the command should work properly. If the command is not recognizable or target out of range, an error message is sent back to alert the user.
2. The car must regulate itself around the target. If the actual distance is closer to the wall, the reverse direction LED lids. If the actual distance is further away, the forward LED lids. The brightness of LEDs are proportional to the required speed to move to the target in exactly one second. Note that for this simple project, the car cannot actually move, so we adjust the actual distance on the sensor manually while running the program. The status and brightness of LEDs must change according to the target and actual position. LEDs must be connected in series with resistors (1K Ohms practically) to limit current.
3. The SSD1306 OLED panel must display the following information
* target distance from wall (cm)  
* actual distance from wall (cm)
* car direction : forward or reverse
* car speed : compute the speed to move to the target position in exactly 1 second. Unit in m/s (meters per second). 
4. You must implement one additional command "name". When a user issues this command, your name-surname, student ID, *and one random number* is outputted to the serial monitor. 

In addition, for this homework, we want to make it an IoT device

5. Using NETPIE IoT cloud only, write these values to NETPIE shadow
* target distance (cm)
* actual distance (cm)
* car speed (m/s)

No other variable should be written to the shadow.

6. Create a NETPIE dashboard that contains these widgets
* a gauge to display actual distance value from shadow. Range must be specified between 50 - 300 cm.
* a slider to set the target distance. Range must be specified between 50 - 300 cm. It should be auto-updated by the value from shadow.
* a guage to display car speed value from shadow. Range must be specified between 0 and maximum speed (that you should compute from the allowable range of the car).
* Two indicator lights for forward and reverse move. These are digital outputs; i.e, either of the lights turns off/on according to the direction of the car. Note that the direction should not be written to NETPIE shadow, so it must be sent directly from the device to dashboard.
* A button to send "name" command to ESP32. Whenever this button is clicked, the name command is sent.
* A text widget to display your name-surname, student ID, and the random number. This information must be sent directly to dashboard from the device. It should not be written to NETPIE shadow.
